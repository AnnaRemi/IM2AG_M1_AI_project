```python
# Load the uploaded CSV files
accuracies_path = '/content/drive/MyDrive/NN Project/results/spiral data/accuracies_64nn_1lrSGD_10000ep.csv'
losses_path = '/content/drive/MyDrive/NN Project/results/spiral data/losses_64nn_1lrSGD_10000ep.csv'
predictions_path = '/content/drive/MyDrive/NN Project/results/spiral data/predictions_64nn_1lrSGD_10000ep.csv'

#vertical data
#true_data_path = '/content/drive/MyDrive/NN Project/data/vertical_data.csv'

#spiral data
true_data_path = '/content/drive/MyDrive/NN Project/data/spiral_data.csv'

# Read the files into pandas DataFrames
accuracies_df = pd.read_csv(accuracies_path)
losses_df = pd.read_csv(losses_path)
predictions_df = pd.read_csv(predictions_path)
last_row_df = predictions_df.tail(1).squeeze().reset_index(drop=True)
true_data_df = pd.read_csv(true_data_path)
assert len(true_data_df) == len(last_row_df), "Lengths of true_data_df and last_row_df are not equal"
```

```python
# Example true labels and predictions (binary classification)
true_data = true_data_df['y']  # True labels
predictions = last_row_df  # Predicted probabilities
```

```python
# Calculate accuracy
accuracy = accuracy_score(true_data, [1 if p > 0.5 else 0 for p in predictions])
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Calculate class-wise accuracy
for class_label in np.unique(true_data):
    class_indices = np.where(true_data == class_label)[0]
    class_accuracy = accuracy_score(true_data[class_indices], predictions[class_indices])
    print(f"Accuracy for class {class_label}: {class_accuracy}")
```

```python
# Calculate confusion matrix
cm = confusion_matrix(true_data, predictions)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=np.unique(true_data), yticklabels=np.unique(true_data))
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.savefig('confmatrix_64nn_1lrSGD_10000ep.png')
plt.show()
```

Explanation: A confusion matrix visualizes the performance of a classification model by showing the counts of true positives, true negatives, false positives, and false negatives for each class.

```python
import numpy as np

def predict_color(point, true_data_df, predictions):
  """Predicts the color of a point based on the model's predictions.

  Args:
    point: A tuple or list representing the (x, y) coordinates of the point.
    true_data_df: The pandas DataFrame containing the true data.
    predictions: The pandas Series containing the model's predictions.

  Returns:
    The predicted color ('red' or 'blue') of the point.
  """

  # Calculate distances between the point and all data points
  distances = np.sqrt(np.sum((true_data_df[['X/0', 'X/1']].values - point)**2, axis=1))

  # Find the index of the nearest data point
  nearest_index = np.argmin(distances)

  # Get the predicted probability for the nearest data point
  predicted_color = predictions[nearest_index]

  # Assign color based on predicted probability
  if predicted_color == 0:
    color = 'blue'
  elif predicted_color == 1:
    color = 'red'
  elif predicted_color == 2:
    color = 'green'

  return color

# Example usage:
x1 = float(input("Enter the first coordinate of the point [-1,1]: "))
x2 = float(input("Enter the second coordinate of the point [-1,1]: ")) # Example point
predicted_color = predict_color((x1,x2), true_data_df, predictions)
print(f"The predicted color for point {(x1,x2)} is {predicted_color}")

#Visualizing the given point over data
plt.scatter(true_data_df['X/0'], true_data_df['X/1'], c=true_data, s=40, cmap='brg')
plt.scatter(x1, x2, c=predicted_color, marker='*', s=100)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Prediction of point over data')
plt.show()
```
